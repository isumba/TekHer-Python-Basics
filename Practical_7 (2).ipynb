### Exercise 39:
Create a filled contour plot for the function `z = x^2 + y^2` over a grid of `x` and `y` values.

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Start', 'Revenue', 'Cost of Goods Sold', 'Operating Expense', 'Profit']
values = [1000, 300, -200, -150, 0]

values[-1] = values[0] + sum(values[1:-1])

cumulative = [values[0]]
for val in values[1:-1]:
    cumulative.append(cumulative[-1] + val)

colors = []
for val in values[1:-1]:
    colors.append('green' if val >= 0 else 'yellow')
colors = ['blue'] + colors + ['magenta']

bar_starts = [0]
for i in range(1, len(values) - 1):
    bar_starts.append(cumulative[i - 1])
bar_starts.append(0)

fig, ax = plt.subplots(figsize=(10, 6))
for i in range(len(values)):
    ax.bar(categories[i], values[i], bottom=bar_starts[i], color=colors[i])

for i in range(1, len(values) - 1):
    y = bar_starts[i]
    ax.plot([i - 1, i], [bar_starts[i - 1] + values[i - 1], y], color='black', linewidth=1)

ax.set_ylabel("Value")
ax.set_title("Waterfall Chart Example")
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Exercise 47:
Create a time series plot using Matplotlib and plot random data over a range of dates.


```python
# Exercise 47
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dates = pd.date_range(start="2024-01-01", periods=30, freq="D")

np.random.seed(0)
values = np.random.randn(30).cumsum()


plt.figure(figsize=(10, 6))
plt.plot(dates, values, marker='o', linestyle='-', color='tab:blue')

plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Plot of Random Data")
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
